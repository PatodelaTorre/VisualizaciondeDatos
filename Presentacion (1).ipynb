```python
import seaborn as sns # For plotting
import pandas as pd
import matplotlib.pyplot as plt # For showing plots

# Load in the data set
df = pd.read_csv('fifa21_male2.csv')
df
```

```python
df.columns
```

```python
sns.boxplot(x = df["Age"], 
            whis=(0, 100),
           ).set_title("Box plot of peso")

plt.show()
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Nationality'],values = ['OVA'],aggfunc = 'median',
                        margins = True
                         )
            
my_pivot.plot()
```

```python
my_pivot=pd.pivot_table(df,
                        index=['Nationality'],
                        values=['OVA','Age'],
                        aggfunc='median',
                        margins=True
                       )
my_pivot
```
