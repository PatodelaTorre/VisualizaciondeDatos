Salary = b0 + b1*Age

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
```

```python
modelA = sm.OLS.from_formula('Salary ~ Age', data=df)
resultA = modelA.fit()
resultA.summary()
```

### Obsevaciones
* Variable dependiente: Salary
* Variable independiente: Age
* R-squared: 0.380; por lo tanto, este modelo no es tan eficiente. Porque explica solo 38% salary basado ...
* Df Model:	1	 
* intercept ... > 0.05, no es estadisticamente significativo

`Salary = 349.4458 + 48.0932*Age`

```python
print('Salary of the person whose age Age is 33: \t', 349.4458 + 48.0932*33)
df.iloc[:2,:]
```

```python
print('Diferencia es \t:', 1500-1936.5214)
print('Diferencia es \t:', 1200-1936.5214)
resultA.resid
```

```python
plt.figure(figsize = (15,10))
sns.set(font_scale = 1.7)

reg_plot = sns.regplot( x = 'Age',
            y = 'Salary',
            data = df,
           marker = '+',
           color = 'green',
                      scatter_kws = {'s':200}
                      )
reg_plot.set_title('Regression plot of Salary vs Age')
reg_plot.text(40, 1500, 'Salary = 349.4458 + 48.0932*Age',
              color = 'green',
              weight = 'semibold',
              size = 'medium'
             )
```

```python
RESIA = resultA.resid
RESIA
```

```python
resi_plot = sns.scatterplot(x = resultA.fittedvalues,
                            y = RESIA)
```

# 2. Pregunta de investigacion 2:
### Cual es el impacto de Age y Gender en Salary
Salary = b0 + b1*Age + b2*Gender

```python
modelB = sm.OLS.from_formula('Salary ~ Age + Gender', data=df)
resultB = modelB.fit()
resultB.summary()
```

### Obsevaciones
* Variable dependiente: Salary
* Variable independiente: Age y Gender
* R-squared: 0.408; por lo tanto, este modelo no es tan eficiente. Porque explica solo 38% salary basado ...
* Df Model:	2	 
* intercept ... > 0.05, no es estadisticamente significativo

`Salary = 749.8157 + 34.2211*Age + 306.7169*Gender`

### Case study
Df Model: 2

### Caso 1:
* Age = 33
* Male = True

```python
print('Salary de una persona con Age 33, Male \t', 749.8157 + 34.2211*33 + 306.7169*1)
```

### Caso 2:
* Age = 33
* Male = False

```python
print('Salary de una persona con Age 33, Female \t', 749.8157 + 34.2211*33 + 306.7169*0)
df.iloc[:2,:]
```

```python
resultB.resid
```
