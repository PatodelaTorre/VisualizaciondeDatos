```python
import seaborn as sns # For plotting
import pandas as pd
import matplotlib.pyplot as plt # For showing plots

# Load in the data set
df = pd.read_csv('fifa21_male2.csv')
df
```

```python
df.columns
```

```python
#EN este caso solamente utilizare las variables 'ID', 'Name', 'Age', 'OVA', 'Nationality', 'Club', 'Position', y las variables 'Wage', 'Value', 'Weight' y 'Height', ya que las demmas no aportan un valor reral a esta investigacion.
```

```python
df.drop(['POT', 'Team & Contract', 'foot',
       'Growth', 'Joined', 'Loan Date End', 'Release Clause',
       'Contract', 'Attacking', 'Crossing', 'Finishing', 'Heading Accuracy',
       'Short Passing', 'Volleys', 'Skill', 'Dribbling', 'Curve',
       'FK Accuracy', 'Long Passing', 'Ball Control', 'Movement',
       'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance',
       'Power', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots',
       'Mentality', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Defending', 'Marking', 'Standing Tackle',
       'Sliding Tackle', 'Goalkeeping', 'GK Diving', 'GK Handling',
       'GK Kicking', 'GK Positioning', 'GK Reflexes', 'Total Stats',
       'Base Stats', 'W/F', 'SM', 'A/W', 'D/W', 'IR', 'PAC', 'SHO', 'PAS',
       'DRI', 'DEF', 'PHY', 'Hits', 'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF',
       'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB',
       'LDM','Player Photo',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK', 'Gender', 'BOV', 'BP','Club Logo', 'Flag Photo'],
       inplace = True,
        axis = 1
       )
df.columns
```

```python
#En este caso terminamos con 11 variables, las cuales seran utilizadas para las preguntas posteriores.
##Hay 6 variables numericas:
# ID es e numero de identificacion que se le da en el juego a ada jugador
# Age es la edad
# OVA es la calificacion promedio del jugador
# Height es la estatura
# Weight es el peso en libras
# Value es el valor del jugador 
# Wage es el salario del jugador 

#Por el otro lado hay 4 variables categoricas
# Name es el nombre
# Nationality la nacionalidad
# Clube es el club en el que juegan
# Position es la posicion 
```

```python
##ALgunos hallazgos interesante que descubri al analizar esos datos son ls siguiente:
# La edad mediana de los jugadores en el FIFA 2021 es de 25 anios
# La mayoria de los jugaodres tiene un promedio de entre 60 y 70 puntos de calificacion
# Existe una correlacion positiva fuerte entre la edad de los jugadores y su calificacion promedio
```

```python
sns.boxplot(x = df["Age"], 
            whis=(0, 100),
           ).set_title("Box plot of peso")

plt.show()
```

```python
sns.histplot(df["OVA"], 
             kde = False).set_title("Histogram of OVA")

# kernel density estimation  KDE

plt.show()
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Nationality'],
                        values = ['OVA'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'mean',
                      
                        margins = True
                         )
            
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Nationality'],values = ['OVA'],aggfunc = 'mean',
                        margins = True
                         )
            
my_pivot.plot()
```

```python
sns.lmplot(x='Age', 
           y='OVA', 
           data= df
          )
plt.title('Scatterplot Position vs Value')
plt.show()
```
